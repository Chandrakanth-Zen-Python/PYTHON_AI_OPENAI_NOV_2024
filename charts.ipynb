```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.DataFrame().from_dict([
                                {'Tool':"AWS Tools","Cost (USD)":300},
                                {'Tool':"Splunk","Cost (USD)":700}
                                ])
```

```python
plt.bar(data['Tool'], data['Cost (USD)'], color=['grey','black'])
plt.title('Cost Comparison Between AWS Tools Vs Splunk', fontsize=16)
plt.xlabel('Tools', fontsize=12)
plt.ylabel('Cost (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
